```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.arange(-50, 50, 0.01)
# ガウス分布の確率密度関数,平均 0、標準偏差1
y = 0.5 * norm.pdf(x, -6, 4) + 0.35 * norm.pdf(x, 8, 2) 
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)


ax.tick_params(
    which="both",
    width=1.5,
    length=8,
    size=10,
    direction="out",
    left=False,
    labelleft=False,
)
ax.xaxis.set_tick_params(labelsize=30, pad=7)
# ax.yaxis.set_tick_params(labelsize=30, pad=7)

ax.plot(x, y, color="black", lw=5)
ax.set_xlim(-15, 15)
plt.savefig("GaussianMixture.png")
```
